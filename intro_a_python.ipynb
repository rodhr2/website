##**Comentarios en Python**

También se pueden agregar **comentarios**, el cual, al momento de ejecutar algún bloque de código, **este no repercute ni se muestra en la consola**. Es importante acotar que **un comentario no es igual a una cadena de caracteres**.

Para poder crearlas, es tan sencillo como solo poner el **signo de gato (#)** y delante de este mismo solo colocar el texto a comentar.

```python
#Comentario 1
#Comentario 2
#Comentario 3

# Esto va a imprimir en pantalla un saludo
```

También se pueden colocar comentarios extensos donde su estructura involucre saltos de linea y mayor contenido.

Para ello, se hará uso de la **tiple comilla simple (''')** al inicio y al final del texto a comentar, o también se puede ocupar de la misma manera la **tiple comilla doble(""")**

```python
print("""
Hola
como
estas?
""")
```

# **Variables, tipo de datos y operaciones aritméticas básicas**

Una **variable** es un **espacio de memoria** reservada para guardar valores y puede cambiar durante la ejecución del programa. Se usa el signo **igual (=)** para crear asignaciones de variables.

Una buena práctica dentro de la programación es guardar cualquier dato dentro de una variable, donde esta misma, tenga un nombre asociado a lo que se está guardando dentro de ella.

- Los nombres de las variables sólo pueden contener letras, números y guiones bajos.
Pueden empezar por una letra o un guión bajo, pero no por un número.
Por ejemplo, puede llamar a una variable `mensaje_1` pero no `1_mensaje`.
- Los espacios no están permitidos en los nombres de variables, pero los guiones bajos pueden utilizarse para separar palabras en los nombres de variables.
para separar palabras en los nombres de las variables. Por ejemplo, `saludo_mensaje` funciona
pero `mensaje saludo` provocará errores.
- Evite utilizar palabras clave de Python y nombres de funciones como nombres de variables;
es decir, no utilice palabras que Python haya reservado para un propósito programático particular, como la palabra `print`.
- Los nombres de las variables deben ser cortos pero descriptivos. Por ejemplo, `nombre` es
mejor que `n`, nombre_alumno es mejor que `s_n`, y `longitud_del_nombre` es mejor
que `longitud_del_nombre_de_personas`.
- Tenga cuidado al utilizar la letra `l` minúscula y la letra `O` mayúscula
porque podrían confundirse con los números `1` y `0`.

En Python **no se declara el tipo de dato**, ya que la variable se adapta según al valor guardado, y así, **automáticamente define su tipo de dato**.

Para conocer el tipo de dato de una variable, se utilizará la función **type()**, y adentro de los parentesis, la variable a consultar.
